```python
# plot mean function
def plot_data_mean(data_dict, contracts, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    ax1 = grid()
    hold(True)
    for i_contract in contracts:
        #convert to dataframe
        df_data = pd.DataFrame(data = data_dict[i_contract], index = index_dates, columns = column_times)
        df_data.ix[plot_date_st:plot_date_ed].mean(axis = 0).T.plot(ax = ax1, label = i_contract)
    for i in range(1,day_end - day_st):
        axvline(x = i* time_int_n-1, color = 'k')
    legend(loc = 'best') 
    return ax1
```

```python
def plot_pos_data(data_pos, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    #convert to dataframe
    df_data = pd.DataFrame(data = data_pos, index = index_dates, columns = column_times)
    ax1 = df_data.ix[plot_date_st:plot_date_ed].T.plot(grid = True)
    for i in range(1,day_end - day_st):
        axvline(x= i* time_int_n, color = 'k')
    legend(loc = 'best')
    return ax1
```

```python
def Event_pl_pos_output(Assets, data, col_name, filename):
    # save data to filename csv file
    df_data = pd.DataFrame()
    col_list= [x + y for x in Assets for y in col_name]
    for i in range(len(Assets)):
        df_data = pd.concat([df_data, data[i]], axis = 1)
    df_data.columns = col_list
    df_data.to_csv(filename)    
```

```python
US = pd.read_csv(inputcsvdir +'//'+ 'FED DATES.csv', index_col= 0, parse_dates= True, header = 1, infer_datetime_format= True)
US = US['2005':'2014-02-01']
US['Date'] = [x.date() for x in US.index]
US['Month'] = [x.month for x in US.index]
US['Minutes Date'] = [x + datetime.timedelta(21) for x in US['Date']]
```

```python
#Events_dates = set(US['Date'])
Events_dates = set.union(set(US['Date']), set(US['Minutes Date'][:-1]))
Events_dates = [x + datetime.timedelta(1) if x in [datetime.date(2007,1,2), datetime.date(2008, 1, 1)] else x for x in Events_dates ]
Events_dates = list((Events_dates))
Events_dates = sort(Events_dates)
Week_days = [i_date.weekday() for i_date in Events_dates]
Event_date_index = pd.DatetimeIndex(Events_dates)
df_week_days = pd.DataFrame(Week_days, index = Event_date_index, columns= ['Week of the day'])
len_events = len(Events_dates)
len_events
```

```python
pos_curve_dict = {}
ewma_seedperiod = 8
window_size = 8
smooth_window = 24.0
mean = np.concatenate((np.diff((event_data_vol_adj_dict['TY']) ,n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['TY'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['TY'] = 1.0 * pos_1 + 0.0 * pos_2

mean = np.concatenate((np.diff((event_data_vol_adj_dict['ES']) ,n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['ES'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['ES'] = 1.0 * pos_1 + 0.0 * pos_2
```
