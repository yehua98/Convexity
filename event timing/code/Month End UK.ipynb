```python
#UK month end
contract = 'G'
#set intraday time points
t_index =pd.date_range('2013-01-01 08:06', '2013-01-01 17:11', freq = '5min') 
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
#load intraday variance
df_variance_profile =  pd.read_csv(inputcsvdir + '\\'+ contract +  '_intraday_variance_5min.csv', index_col=0)

#load ATR 
df_ATR = pd.read_csv(inputcsvdir + '\\'+ contract+ '_ATR.csv',index_col=0, parse_dates=True)
dates = list(df_ATR.index.date)
#load 5 mins data for all trading days
df_data_all = pd.read_csv(inputcsvdir + '\\'+ contract+ '_5min.csv', index_col= 0, parse_dates= True)  
```

```python
pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
ME_dates = df_ATR.index[df_ATR.index.month[0:-1] - df_ATR.index.month[1:] != 0]
```

```python
def get_event_data(day_st, day_end, time_interval, Events_dates, df_data_all, df_ATR, dates):   
    event_data = np.empty((len(Events_dates), (day_end-day_st)*len(time_interval)))
    #event_data_times = np.tile(time_interval,(event_n, day_end - day_st))
    event_n = len(Events_dates)
    event_data_date_time = np.tile(Events_dates[0], (event_n, (day_end-day_st)*len(time_interval)))
    event_vol = np.empty(len(Events_dates))
    for i,date_i in enumerate(Events_dates):
        # get location on dates, note dates are the index of df_ATR
        date_i = date_i.date()
        idx = dates.index(date_i)
        event_vol[i] = df_ATR.ewma[dates[idx + day_st -1]]  # day_start-1 as the vol
        for j, value in enumerate(range(day_st, day_end)):
           idx_tmp = idx + value
           df_tmp = et.get_data_by_date(df_data_all,dates,dates[idx_tmp])
           event_data[i, j*time_int_n:(j+1)*time_int_n]= df_tmp.values.T
           event_data_date_time[i, j*time_int_n:(j+1)*time_int_n] = list(df_tmp.index)

    # re-center end of T-1 day as zeros
    event_data = event_data[:, :] - event_data[:,(day_end-day_st)/2*time_int_n-1][:, np.newaxis]
    event_data_vol_adj = event_data /event_vol[:, np.newaxis]
    return event_data, event_vol, event_data_vol_adj, event_data_date_time 
```

```python
day_st = -2
day_end = 2
event_data, event_vol, event_data_vol_adj, event_data_date_time = get_event_data(day_st = day_st, day_end = day_end, time_interval= time_interval, Events_dates = ME_dates, df_data_all = df_data_all, df_ATR = df_ATR, dates = dates)
```

```python
# UK FI
#set window parameters
ewma_seedperiod = 12
window_size = 12

#set the events
len_events = len(ME_dates)

df_data_vol_adj = pd.DataFrame(event_data_vol_adj)
df_data =pd.DataFrame(event_data)
data_date_time = event_data_date_time

A = np.concatenate((np.diff(df_data_vol_adj.values,n=1,axis = 1), np.zeros((len_events, 1))), 1)
B = np.tile(df_variance_profile.values.T,(len_events,day_end - day_st))
pos_raw = np.divide(A, B)
pos_curve_1= np.zeros(pos_raw.shape)     # expanding windwow position
pos_curve_2= np.zeros(pos_raw.shape)     # rolling window position
x1 = arange(pos_raw.shape[1])
smooth_window = 36.0

#rolling window kernal
for i_row in arange(ewma_seedperiod, pos_curve_2.shape[0]):
    X = np.tile(x1, window_size)
    Y = pos_raw[i_row-window_size+1:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1, smooth_window)
    pos_curve_2[i_row, :] = f_x1/var_f_x1

# use weighted kernal smoothing
#for i_row in arange(ewma_seedperiod, pos_curve_2.shape[0]):
#    X = np.tile(x1, window_size)
#    Y = pos_raw[i_row-window_size+1:i_row+1, :].flatten()
#    ws = np.array(et.gen_weights(1.0/window_size, window_size))
#    weights = np.tile(ws[ ::-1], (pos_curve_2.shape[1],1)).T.flatten()
#    f_x1, var_f_x1 = et.calc_gaussian_kernel_regression_weighted(X,Y,weights, x1, smooth_window)
#    pos_curve_2[i_row, :] = f_x1/var_f_x1
    
#expanding window kernal
for i_row in arange(0, pos_curve_1.shape[0]):
    X = np.tile(x1, i_row+1)
    Y = pos_raw[:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1,smooth_window)
    pos_curve_1[i_row, :] = f_x1/var_f_x1  
    
# weighted kernal smoothing
#for i_row in arange(0, pos_curve_1.shape[0]):
#    X = np.tile(x1, i_row+1)
#    Y = pos_raw[:i_row+1, :].flatten()
#    ws = np.array(et.gen_weights(1.0/(i_row+1), i_row+1))
#    weights = np.tile(ws[ ::-1], (pos_curve_2.shape[1],1)).T.flatten()
#    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1,smooth_window)
#    pos_curve_1[i_row, :] = f_x1/var_f_x1
```

```python
rcParams['figure.figsize'] = 15, 4
pos_curve_1[:, 3*time_int_n:] = 0
pos_curve_2[:, 3*time_int_n:] = 0
pos_curve = 2.5*pos_curve_1 + 1.0*pos_curve_2    # note the pos_cruve magitute are different. Should weight by IR

ME_dates_str = [idate.strftime("%Y-%m-%d") for idate in ME_dates]
time_interval_str = [itime.strftime("%H:%M") for itime in time_interval*(day_end -day_st)]

#convert pos to dataframe
df_pos_curve = pd.DataFrame(data = pos_curve, index = ME_dates, columns=time_interval_str)
df_pos_curve_1 = pd.DataFrame(data = pos_curve_1, index = ME_dates, columns=time_interval_str)
df_pos_curve_2 = pd.DataFrame(data = pos_curve_2, index = ME_dates, columns=time_interval_str)

df_pos_curve['2013':].T.plot()
#df_pos_curve.ix['2012-01-31'].T.plot()
for i in range(1,day_end - day_st):
    axvline(x= i*time_int_n-1, color = 'k')
legend(loc = 'best')
df_pos_curve_1['2013'].T.plot()
#df_pos_curve.ix['2012-01-31'].T.plot()
for i in range(1,day_end - day_st):
    axvline(x= i*time_int_n-1, color = 'k')
legend(loc = 'best')
df_pos_curve_2['2013'].T.plot()
#df_pos_curve.ix['2012-01-31'].T.plot()
for i in range(1,day_end - day_st):
    axvline(x= i*time_int_n-1, color = 'k')
legend(loc = 'best')
```

```python
#PL calculation
FS = 1000000
gearing = 20
cost_const = 0.5*0.01

df_pos_curve = pd.DataFrame(pos_curve, index = df_data.index)
s_vol = pd.Series(data = event_vol, index= df_data.index)
df_pos = df_pos_curve.shift(1).div(s_vol, axis = 0)*FS/gearing
pl = df_pos.values*df_data.T.diff(1).T.values
cost = cost_const*df_pos.T.diff(1).T.fillna(value = 0).abs().values
df_pl = pd.DataFrame({'pl':pl.flatten(), 'pl_net':(pl-cost).flatten()},index = data_date_time.flatten())

# convert to daily PL
daily_pl = df_pl.groupby(lambda x: x.date()).sum()    
daily_pl.cumsum().plot()

#PL analysis
btest = btestlib.Backtest(params)
calendar_dates = pd.date_range(start = daily_pl.index[0], end = daily_pl.index[-1], freq='B')
IR = btest.calc_IR(daily_pl.reindex(index = calendar_dates, fill_value= 0), annual_factor= 252)
hit_ratio = et.get_hit_ratio(pl)
hit_ratio_net = et.get_hit_ratio(pl - cost)
```
