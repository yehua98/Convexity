```python
#get daily ATR by intraday data
def get_ATR(data_df, open_time, close_time):
    df = data_df[['Open', 'High', 'Low', 'Close']]
    df = df.between_time(open_time, close_time)
    trading_days = np.unique(df.index.date)
    trading_days = sorted(trading_days)
    #get daily OHLC
    df_days_OHLC = pd.DataFrame(np.nan, index = trading_days, columns=df.columns) 
    for i_day in trading_days:
        df_daily_data = df[i_day.strftime('%Y-%m-%d')]
        df_days_OHLC.ix[i_day, 'Open'] = df_daily_data.Open[0]
        df_days_OHLC.ix[i_day, 'High'] = np.max(df_daily_data.values)
        df_days_OHLC.ix[i_day, 'Low'] = np.min(df_daily_data.values)
        df_days_OHLC.ix[i_day, 'Close'] = df_daily_data.Close[-1]
    
    df_ATR = pd.DataFrame(np.nan, index = trading_days, columns=['ATR'])
    OHLC = df_days_OHLC.values
    
    #get daily ATR
    df_ATR.ATR[0] = OHLC[0, 1] - OHLC[0, 2]
    for i, i_day in enumerate(trading_days[1:], start = 1):
        df_ATR.ix[i_day, 'ATR'] = max([OHLC[i, 1]- OHLC[i, 2], abs(OHLC[i, 1] - OHLC[i-1, 3]),abs(OHLC[i, 2] - OHLC[i-1, 3])]) 
    
    #get smoothed ATR through EWMA
    df_ATR['ewma'] = pd.ewma(df_ATR.ATR, com = 2, min_periods=3)
    return df_ATR    
```

```python
#set the events and obtain position function

#set window parameters
ewma_seedperiod = 20
window_size = 20

#set the events
event_idx = (UK.IIB ==0)& (UK.Tenor>=5)
len_events = len(Events_dates[event_idx])

df_data_vol_adj =pd.DataFrame(event_data_vol_adj[event_idx.values, :])
df_data =pd.DataFrame(event_data[event_idx.values, :]) 

A = np.concatenate((np.diff(df_data_vol_adj.values,n=1,axis = 1), np.zeros((len_events, 1))), 1)
B = np.tile(variance_profile,(len_events,day_end - day_st))
pos_raw = np.divide(A, B)
pos_curve_1= np.zeros(pos_raw.shape)     # expanding windwow position
pos_curve_2= np.zeros(pos_raw.shape)     # rolling window position
x1 = arange(pos_raw.shape[1])
smooth_window = 18.0

#rolling window kernal
for i_row in arange(ewma_seedperiod, pos_curve_2.shape[0]):
    X = np.tile(x1, window_size)
    Y = pos_raw[i_row-window_size+1:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1, smooth_window)
    pos_curve_2[i_row, :] = f_x1/var_f_x1

    
#expanding window kernal
for i_row in arange(0, pos_curve_1.shape[0]):
    X = np.tile(x1, i_row+1)
    Y = pos_raw[:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1, smooth_window)
    pos_curve_1[i_row, :] = f_x1/var_f_x1

#remove unrelated pos, 110:T-1 datetime.time(8, 6), 300: T datetime.time(14, 46)
pos_curve_1[:, 0:110] = 0
pos_curve_1[:, 300:] = 0
pos_curve_2[:, 0:110] = 0
pos_curve_2[:, 300:] = 0

pos_curve = 0.5*pos_curve_1 + 0.5*pos_curve_2
```
