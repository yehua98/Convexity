```python
set_printoptions(threshold=10000)
import pandas as pd
import numpy as np
import os
import sys
import ConfigParser
from pandas import datetime
from datetime import time, date

# read the configuration file and set the path to IDPResearch directory
config = ConfigParser.RawConfigParser()
config.read(os.path.abspath('E:\github\SystematicStrategies\idpresearch.cfg'))
sys.path.append(os.path.abspath(config.get('IDPResearch','path')))

from idpresearch import *
```

```python
inputcsvdir = os.path.abspath('../data/US/')
outputcsvdir = os.path.abspath('../data/')
df= pd.read_csv(inputcsvdir + '\\US.csv')
df = df[['Date Time', 'Event']]
df['Date Time']= pd.to_datetime(df['Date Time'])
df = df.set_index(['Date Time'])
df['Event'] = df['Event'].map(lambda x: x.lower())
```

```python
def parse_to_get_size(s, ccy):  
    if ("bln" in s) or ("billion" in s):
        s = s.replace('billion', 'bln')
        y = float(s.split(" bln")[0].split(ccy)[1].replace(' ', ''))
    elif ("mln" in s) or ("million" in s):
        s = s.replace('million', 'mln')
        y = float(s.split(" mln")[0].split(ccy)[1].replace(' ', '')) / 1000
    else:
        y = np.nan
    return y
def parse_to_get_yield(s):
    if ("yld" in s):
        y = float(s.split("yld ")[1].split("%")[0].replace(' ', '')) / 100
    elif ("yield" in s):
        y = float(s.split("yield ")[1].split("%")[0].replace(' ', '')) / 100
    else:
        y = np.nan
    return y
def parse_to_get_tenor(s):
    if ("-year" in s) :
        y = int(s.split("-year ")[0].split(" ")[-1])
    elif ("year" in s):
        s = s.replace(' year', '-year')
        y = int(s.split("-year ")[0].split(" ")[-1])
    else:
        y = np.nan
    return y
def parse_to_get_iib(s):
    if ("tips" in s) or ("link" in s) or ('frn' in s):
        y = 1
    else:
        y = 0
    return y
```

```python
for i in range(len(df)):
    parse_to_get_tenor(df['Event'].ix[i])
```

```python
df['Tenor'] = df['Event'].map(parse_to_get_tenor)
df['IIB'] = df['Event'].map(parse_to_get_iib)
df['Yield'] = df['Event'].map(parse_to_get_yield)
df['Size'] = df['Event'].apply(parse_to_get_size, args=("usd",))
df['Bid/Cover'] = np.nan
df['Tail']= np.nan
df = df[['Event', 'IIB', 'Tenor', 'Yield', 'Size', 'Bid/Cover', 'Tail']]
df.tail(n = 10)
```
