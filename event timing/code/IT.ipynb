```python
# plot mean function
def plot_data_mean(data_dict, contracts, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    ax1 = grid()
    hold(True)
    for i_contract in contracts:
        #convert to dataframe
        df_data = pd.DataFrame(data = data_dict[i_contract], index = index_dates, columns = column_times)
        df_data.ix[plot_date_st:plot_date_ed].mean(axis = 0).T.plot(ax = ax1, label = i_contract)
    for i in range(1,day_end - day_st):
        axvline(x = i* time_int_n-1, color = 'k')
    legend(loc = 'best') 
    return ax1
```

```python
def plot_pos_data(data_pos, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    #convert to dataframe
    df_data = pd.DataFrame(data = data_pos, index = index_dates, columns = column_times)
    ax1 = df_data.ix[plot_date_st:plot_date_ed].T.plot(grid = True)
    for i in range(1,day_end - day_st):
        axvline(x= i* time_int_n, color = 'k')
    legend(loc = 'best')
    return ax1
```

```python
def Event_pl_pos_output(Assets, data, col_name, filename):
    # save data to filename csv file
    df_data = pd.DataFrame()
    col_list= [x + y for x in Assets for y in col_name]
    for i in range(len(Assets)):
        df_data = pd.concat([df_data, data[i]], axis = 1)
    df_data.columns = col_list
    df_data.to_csv(filename)    
```

```python
Events_dates_5y = set(IT['Date'].ix[(IT['Tenor'] < 7).values & (IT['Tenor'] > 2).values])
Events_dates_10y = set(IT['Date'].ix[(IT['Tenor'] >= 7).values]) 
Events_dates = Events_dates_5y  - Events_dates_10y
Events_dates = list(Events_dates)
Events_dates = sort(Events_dates)
Week_days = [i_date.weekday() for i_date in Events_dates]
Event_date_index = pd.DatetimeIndex(Events_dates)
df_week_days = pd.DataFrame(Week_days, index = Event_date_index, columns= ['Week of the day'])
len_events = len(Events_dates)
len_events
```

```python
#load main event data
contracts = ['IK', 'OE']
t_index = pd.date_range('2013-01-01 08:06', '2013-01-01 18:11', freq = '5min') 
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
```

```python
event_vol_dict = {}
event_data_dict = {}
event_data_vol_adj_dict = {}
df_variance_profile = {}

#load intraday variance
for i_contract in contracts:
    df_variance_profile[i_contract] =  pd.read_csv(inputcsvdir + '\\'+ i_contract+  '_intraday_variance_5min.csv', index_col=0)
    #load ATR 
    df_ATR = pd.read_csv(inputcsvdir + '\\'+ i_contract+ '_ATR.csv',index_col=0, parse_dates=True, infer_datetime_format= True)
    dates = list(df_ATR.index.date)
    #load 5 mins data for all trading days
    df_data_all = pd.read_csv(inputcsvdir + '\\'+ i_contract+ '_5min.csv', index_col= 0, parse_dates= True, infer_datetime_format= True) 
    #df_data_all = df_data_all.tz_localize('US/Central')

    # load data matrix, event_data is a matrix whose rows are the evenet dates and col are intraday time interval 
    day_st  = -3    # date from T-3 day
    day_end = 4        # date end at T+3 day. T+4 is excluded
    event_data = np.empty((len(Events_dates), (day_end-day_st)*len(time_interval)))
    event_data_times = np.tile(time_interval,(len_events, day_end - day_st))
    event_data_date_time = np.tile(Events_dates[0], (len_events, (day_end-day_st)*len(time_interval)))
    event_vol = np.empty(len(Events_dates))
    for i,date_i in enumerate(Events_dates):
        # get location on dates, note dates are the index of df_ATR
        idx = dates.index(date_i)
        event_vol[i] = df_ATR.ewma[dates[idx + day_st -1]]  # day_start-1 as the vol
        for j, value in enumerate(range(day_st, day_end)):
           idx_tmp = idx + value
           df_tmp = et.get_data_by_date(df_data_all,dates,dates[idx_tmp])
           #df_tmp = df_tmp.between_time(start_time = time_interval[0], end_time = time_interval[-1])  #set start and end time     
           event_data[i, j*time_int_n:(j+1)*time_int_n]= df_tmp.values.T
           event_data_date_time[i, j*time_int_n:(j+1)*time_int_n] = list(df_tmp.index)

    # re-center end of T-1 day as zeros
    event_data = event_data[:, :] - event_data[:,(day_end-day_st)/2*time_int_n-1][:, np.newaxis]
    event_data_vol_adj = event_data /event_vol[:, np.newaxis]
    
    event_vol_dict[i_contract] = event_vol
    event_data_dict[i_contract] = event_data
    event_data_vol_adj_dict[i_contract] = event_data_vol_adj  
```

```python
rcParams['figure.figsize'] = 18, 3
grid()
plot(event_data_vol_adj_dict['IK'].mean(axis=0), 'm', label = 'IK')
plot(event_data_vol_adj_dict['OE'].mean(axis=0), 'g', label = 'OE')
for i in range(1,day_end - day_st):
    axvline(x= i*time_int_n-1, color = 'k')
legend(loc = 'best')
```
