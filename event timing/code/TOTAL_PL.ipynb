```python
def plot_daily_pl(df_pl):
    contracts = list(set([x.split(' ')[0] for x in df_pl.columns]))
    pl_list = [x + ' pl' for x in contracts] 
    pl_net_list = [x + ' pl_net' for x in contracts]
    df_pl_daily = df_pl.ix[:, pl_list].sum(axis= 1).resample('B', how = 'sum')
    df_pl_daily_net = df_pl.ix[:, pl_net_list].sum(axis= 1).resample('B', how = 'sum')
    df_pl_daily = df_pl_daily.fillna(value = 0)
    df_pl_daily_net = df_pl_daily_net.fillna(value = 0)
    ax1 = df_pl_daily.cumsum().plot()
    df_pl_daily_net.cumsum().plot(ax = ax1)
```

```python
def risk_output_contract_rtn(df_pl_all, inputcsvdir):
    contracts = list(set([x.split(' ')[0] for x in df_pl_all.columns]))
    for i in range(len(contracts)):
        #load contract price data
        df_contract_price = pd.read_csv(inputcsvdir + '\\'+ contracts[i]+ '_5min.csv',  index_col=0, parse_dates=True, infer_datetime_format= True)
        df_contract_daily = df_contract_price.resample('B', how = 'last')
        df_contract_daily_rtn = df_contract_daily.diff()
        df_contract_daily_rtn.to_csv(inputcsvdir + '\\' + 'Comb_dailyRtn_' + contracts[i] + '.csv')
        df_contract_hourly = df_contract_price.resample('1h', how = 'last')
        df_contract_hourly_rtn = df_contract_hourly.diff()
        df_contract_hourly_rtn.to_csv(inputcsvdir + '\\' + 'Comb_hourlyRtn_' + contracts[i] + '.csv')  
```

```python
def risk_output_pl(df_pl_all, inputcsvdir):
    contracts = list(set([x.split(' ')[0] for x in df_pl_all.columns]))
    pl_list = [x + ' pl' for x in contracts] 
    pl_net_list = [x + ' pl_net' for x in contracts]
    for i in range(len(contracts)):
        df_pl_contract_daily = df_pl_all.ix[:, [pl_list[i], pl_net_list[i]]].resample('B', how = 'sum')
        df_pl_contract_daily.to_csv(inputcsvdir + '\\' + 'Comb_dailyPL_' + contracts[i] + '.csv')
        df_pl_contract_hourly = df_pl_all.ix[:, [pl_list[i], pl_net_list[i]]].resample('1h', how = 'sum')
        df_pl_contract_hourly.to_csv(inputcsvdir + '\\' + 'Comb_hourlyPL_' + contracts[i] + '.csv')    
```

```python
def risk_output_pos(df_pos_all, inputcsvdir):
    contracts = list(set([x.split(' ')[0] for x in df_pos_all.columns]))
    pos_list = [x + ' pos' for x in contracts]
    for i in range(len(contracts)):
        #daily
        df_pos_contract_daily_close = df_pos_all.ix[:, pos_list[i]].resample('B', how = 'last')
        df_pos_contract_daily_max = df_pos_all.ix[:, pos_list[i]].resample('B', how = 'max')
        df_pos_contract_daily_min = df_pos_all.ix[:, pos_list[i]].resample('B', how = 'min')
        df_pos_contract_daily = pd.concat([df_pos_contract_daily_close, df_pos_contract_daily_max, df_pos_contract_daily_min], axis = 1)
        df_pos_contract_daily.columns = ['daily close', 'daily max', 'daily min']
        df_pos_contract_daily.to_csv(inputcsvdir + '\\' + 'Comb_dailyPos_' + contracts[i] + '.csv')
        #hourly
        df_pos_contract_hourly = df_pos_all.ix[:,  pos_list[i]].resample('1h', how = 'last')
        df_pos_contract_hourly.name = 'hourly pos'
        df_pos_contract_hourly.to_csv(inputcsvdir + '\\' + 'Comb_hourlyPos_' + contracts[i] + '.csv')
```

```python
def standarlize_pl(PL):
    std_pl = PL['pl_net'].ix[abs(PL['pl_net'])> 0].mean()
    PL_new = PL/std_pl
    return PL_new
```
