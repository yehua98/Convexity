```python
    def create_roll_adjusted_series(self, df, calendar, method='ratio', method_df=None, method_out=False, column='Generic'):
        '''
        Create a single data series by combining a set of series based on a calendar of roll dates and adjustment method.
        This function can be used to create a historical continuous data series of prices for products that trade as
        short-lived contracts.

        Parameters
        ----------
        df : DataFrame
            Data to combine and adjust.
        calendar : Series
            Roll calendar with dates as the index and columns names as the values.
        method : str, optional
            Method of adjustment. Options available are: 'ratio', 'diff', 'none', and 'amort' (defaults to 'ratio').
        method_df : df, optional
            Predetermined adjustments for the specified method--only applicable for 'ratio' and 'diff' (defaults to None).
        method_out : bool, optional
            If True, will return a DataFrame of the adjustments used based on the method selected.
        column : str, otpional
            Name of the output column (defaults to 'Generic').

        Returns
        -------
        out_df : DataFrame
            Stitched and adjusted series.
        adj_df : DataFrame
            Adjustments used based on the selected series if 'method_out' is True.

        Details
        -------
        The 'ratio' method performs backward ratio-adjustment by dividing the level of the back contract (new front contract)
        by the level of the front contract on the roll date and multiplying the entire historical time series by that ratio.
        This is performed iteratively on each roll date from the start of the time series. If the future is in contango, the
        back contract will be higher than the front contract and the absolute level of the historical series will be artificially
        increased. If the the future is in backwardation, the opposite will occur--the asbolute level of the historical series
        will be reduced. While this method affects the absolute level of the price series, it will preserve the true returns and
        abstract the price return on roll dates from the total return (roll + price).

        The 'diff' method performs backward difference-adjustment. It is similiar to the 'ratio' method except that instead of
        taking the ratio of back-contract/front-contract the difference is computed and applied historically (back-contract
        minus front-contract).

        The 'amort' method...TBA

        The 'none' method performs no adjustment. It merely 'stitches' the individual contracts together based on the roll calendar
        forming a single time series.

        Sources
        -------
        http://www.ipedr.com/vol29/48-CEBMM2012-R00003.pdf
        C:/../IDPResearch/docs/CDS_Roll_Specification.pdf

        '''

        if method not in ['ratio', 'diff', 'none', 'amort']:
            raise ValueError('method is not valid: ' + method)

        if method not in ['none', 'amort'] and method_out:
            adj_df = pd.DataFrame(columns=['Date', 'Adjustment'])

        # Predetermined methods only applicavble for ratio or diff
        if method not in ['ratio', 'diff']:
            method_df = None

        if not isinstance(calendar, pd.Series):
            raise ValueError('calendar must be a pandas Series')

        out_df = pd.DataFrame(np.nan, index=df.index, columns=[column])
        ndim = len(df.columns) - 1
        curr_roll_date = df.index[0]
        before_roll_date = df.index[0]
        for idx, name in enumerate(df):
            roll_date = calendar[calendar == name].index[0]

            # If this is the most current contract available
            if idx >= ndim:
                out_df[column].ix[curr_roll_date:df.icol(idx).index[-1]] = df.icol(idx).ix[curr_roll_date:]
                break

            front_df = df.icol(idx)
            back_df = df.icol(idx + 1)

            if not len(front_df.dropna().index) or not len(back_df.dropna().index):
                continue

            if idx == 0:
                prior_roll_date = df.index[0]
            else:
                prior_roll_date = self.dttm.shift_by_weekdays(before_roll_date, 1)

            curr_roll_date = self.dttm.get_closest_date(roll_date, back_df.dropna().index, direction='forward')

            # Skip roll_date if prior to df index
            if curr_roll_date < df.index[0]:
                continue

            # Exit if the roll_date is later than the available data
            if curr_roll_date > df.index[-1]:
                out_df[column].ix[prior_roll_date:back_df.index[-1]] = back_df.ix[prior_roll_date:]
                break

            before_roll_date = self.dttm.shift_by_weekdays(curr_roll_date, -1)
            before_roll_date = self.dttm.get_closest_date(before_roll_date, front_df.index, direction='backward')

            if method != 'none':
                # Get adjustment
                if not method_df:
                    val_date = self.dttm.get_closest_date(curr_roll_date, front_df.dropna().index, direction='backward')
                    if method == 'ratio':
                        adj = back_df.ix[curr_roll_date] / front_df.ix[val_date]
                    elif method == 'diff':
                        adj = back_df.ix[curr_roll_date] - front_df.ix[val_date]
                    elif method == 'amort':
                        adj = (back_df.ix[curr_roll_date] - front_df.ix[val_date]) / float(len(out_df[column].ix[prior_roll_date:before_roll_date]))

                else:
                    # Need to reconcile potential mismatches between the calendar_df and the method_df
                    curr_roll_date = self.dttm.get_closest_date(curr_roll_date, list(method_df.icol(0)), direction='absolute')
                    adj = method_df[method_df.icol(0) == curr_roll_date].icol(1).iloc[0]

                out_df[column].ix[prior_roll_date:before_roll_date] = front_df.ix[prior_roll_date:before_roll_date]

                # Apply adjustment
                if method == 'ratio':
                    out_df[column].ix[:before_roll_date] = out_df[column].ix[:before_roll_date] * adj
                elif method == 'diff':
                    out_df[column].ix[:before_roll_date] = out_df[column].ix[:before_roll_date] + adj
                elif method == 'amort':
                     temp_df =  out_df.ix[prior_roll_date:before_roll_date].reset_index()
                     temp_df['Adj'] = (temp_df.index + 1) * adj
                     out_df[column].ix[prior_roll_date:before_roll_date] = temp_df[column] + temp_df['Adj']

                if method_out:
                    adj_df = adj_df.append({'Date': curr_roll_date, 'Adjustment': adj}, ignore_index=True)

            # For no adjustment
            else:
                out_df[column].ix[prior_roll_date:before_roll_date] = front_df.ix[prior_roll_date:before_roll_date]

        if method != 'none' and method_out:
            return out_df, adj_df
        else:
            return out_df
```
