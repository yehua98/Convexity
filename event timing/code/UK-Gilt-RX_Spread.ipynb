```python
#load main event data
t_index = pd.date_range('2013-01-01 08:06', '2013-01-01 17:11', freq = '5min') 
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
#load intraday variance
df_variance_profile =  pd.read_csv(inputcsvdir + '\\'+ 'G'+  '_intraday_variance_5min.csv', index_col=0, infer_datetime_format= True)

#load ATR 
df_ATR = pd.read_csv(inputcsvdir + '\\'+ 'G'+ '_ATR.csv',index_col=0, parse_dates=True, infer_datetime_format= True)
dates = list(df_ATR.index.date)
#load 5 mins data for all trading days
df_data_all = pd.read_csv(inputcsvdir + '\\'+ 'G'+ '_5min.csv', index_col= 0, parse_dates= True, infer_datetime_format= True) 
df_data_all = df_data_all.tz_localize('Europe/London')
```

```python
# load data matrix, event_data is a matrix whose rows are the evenet dates and col are intraday time interval 
day_st  = - 2    # date from T-2 day
day_end = 2        # date end at T+1 day. T+2 is excluded
event_data = np.empty((len(Events_dates), (day_end-day_st)*len(time_interval)))
event_data_times = np.tile(time_interval,(event_n, day_end - day_st))
event_data_date_time = np.tile(Events_dates[0], (event_n, (day_end-day_st)*len(time_interval)))
event_vol = np.empty(len(Events_dates))
for i,date_i in enumerate(Events_dates):
    # get location on dates, note dates are the index of df_ATR
    idx = dates.index(date_i)
    event_vol[i] = df_ATR.ewma[dates[idx + day_st -1]]  # day_start-1 as the vol
    for j, value in enumerate(range(day_st, day_end)):
       idx_tmp = idx + value
       df_tmp = et.get_data_by_date(df_data_all,dates,dates[idx_tmp])
       event_data[i, j*time_int_n:(j+1)*time_int_n]= df_tmp.values.T
       event_data_date_time[i, j*time_int_n:(j+1)*time_int_n] = list(df_tmp.index)

# re-center end of T-1 day as zeros
event_data = event_data[:, :] - event_data[:,(day_end-day_st)/2*time_int_n-1][:, np.newaxis]
event_data_vol_adj = event_data /event_vol[:, np.newaxis] 
```

```python
# load hedge data
event_data_hedge = np.empty((len(Events_dates), (day_end-day_st)*len(time_interval)))
event_vol_hedge = np.empty(len(Events_dates))
for i,date_i in enumerate(Events_dates):
    # get location on dates, note dates are the index of df_ATR
    idx = dates.index(date_i)
    event_vol_hedge[i] = df_ATR_hedge.ewma[dates[idx + day_st -1]]  # day_start-1 as the vol
    for j, value in enumerate(range(day_st, day_end)):
       idx_tmp = idx + value
       df_tmp = et.get_data_by_date(df_data_hedge,dates_hedge,dates[idx_tmp])
       event_data_hedge[i, j*time_int_n:(j+1)*time_int_n]= df_tmp.values.T

# re-center end of T-1 day as zeros
event_data_hedge = event_data_hedge[:, :] - event_data_hedge[:,(day_end-day_st)/2*time_int_n-1][:, np.newaxis]
event_data_vol_adj_hedge = event_data_hedge /event_vol_hedge[:, np.newaxis] 
```

```python
# pos calculation
#set window parameters
ewma_seedperiod = 15
window_size = 15

#set the events
len_events = len(Events_dates)

df_data_vol_adj = pd.DataFrame(event_data_vol_adj - event_data_vol_adj_hedge)
data_date_time = event_data_date_time

A = np.concatenate((np.diff(df_data_vol_adj.values,n=1,axis = 1), np.zeros((len_events, 1))), 1)
B = np.tile(df_variance_profile.values.T,(len_events,day_end - day_st))
pos_raw = np.divide(A, B)
pos_curve_1= np.zeros(pos_raw.shape)     # expanding windwow position
pos_curve_2= np.zeros(pos_raw.shape)     # rolling window position
x1 = arange(pos_raw.shape[1])
smooth_window = 30.0

#rolling window kernal
for i_row in arange(ewma_seedperiod, pos_curve_2.shape[0]):
    X = np.tile(x1, window_size)
    Y = pos_raw[i_row-window_size+1:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1, smooth_window)
    pos_curve_2[i_row, :] = f_x1/var_f_x1

    
#expanding window kernal
for i_row in arange(0, pos_curve_1.shape[0]):
    X = np.tile(x1, i_row+1)
    Y = pos_raw[:i_row+1, :].flatten()
    f_x1, var_f_x1 = sig.calc_gaussian_kernel_regression(X,Y,x1, smooth_window)
    pos_curve_1[i_row, :] = f_x1/var_f_x1
```

```python
pos_curve = 0.6*pos_curve_1 + 0.4*pos_curve_2
pos_curve[:, 4*time_int_n-1:] = 0
#pos_curve[:, 360:] = 0
pos_curve[:, :100] = 0
#pos_curve[:, :0*time_int_n-1] = 0
#automate position catch
```

```python
rcParams['figure.figsize'] = 15, 4
pos_curve = 1.5*pos_curve_1 + 1*pos_curve_2
#pos_curve[:, 3*time_int_n:] = 0
#pos_curve[:, 360:] = 0
#pos_curve[:, :100] = 0
#pos_curve[:, :0*time_int_n] = 0
#automate position catch

pos_select = numpy.zeros(pos_curve.shape)
for i_row in arange(0, pos_curve.shape[0]):
    zero_crossings = np.where(np.diff(np.sign(pos_curve[i_row])))[0]
    if len(zero_crossings) >= 3:    # cut between the first and the end points
       pos_select[i_row, zero_crossings[0]: zero_crossings[-1]] = 1.0     
    elif len(zero_crossings) == 1 or len(zero_crossings) == 2:  # take whole range
       pos_select[i_row, zero_crossings[0]] = 1.0        
    else: 
       print "There is no position change in row" + i_row
       pos_select[i_row, :] = 1.0 
        
pos_curve = pos_curve * pos_select 
```

```python
def get_pl_pos(FS, gearing, cost_const, FPV, df_data, pos_curve, event_vol, data_date_time):
    df_pos_curve = pd.DataFrame(pos_curve, index = df_data.index)
    s_vol = pd.Series(data = event_vol, index= df_data.index)
    df_pos = df_pos_curve.shift(1).div(s_vol, axis = 0)*FS/gearing/FPV
    pl = df_pos.values*df_data.T.diff(1).T.values*FPV
    cost = cost_const*df_pos.T.diff(1).T.fillna(value = 0).abs().values*FPV
    df_pl = pd.DataFrame({'pl':pl.flatten(), 'pl_net':(pl-cost).flatten()},index = data_date_time.flatten())

    # convert to daily PL
    daily_pl = df_pl.groupby(lambda x: x.date()).sum()
    daily_pl.index= pd.DatetimeIndex(daily_pl.index)
    calendar_dates = pd.date_range(start = daily_pl.index[0], end = daily_pl.index[-1], freq='B')
    daily_pl = daily_pl.reindex(index = calendar_dates, fill_value= 0)

    #intraday positon to end of day position
    df_pos_intraday = pd.DataFrame({'pos':df_pos.values.flatten()},index = data_date_time.flatten())
    daily_pos = df_pos_intraday.groupby(lambda x: x.date()).apply(lambda x: x.ix[-1,:])
    daily_pos.index= pd.DatetimeIndex(daily_pos.index)
    daily_pos = daily_pos.reindex(index = calendar_dates, fill_value= 0)
   
    return df_pl, daily_pl, df_pos_intraday, daily_pos 
```

```python
#PL calculation
FS = 1000000
gearing = 20
cost_const = 0.25*0.01    
FPV = 1000
df_data = pd.DataFrame(event_data)
df_data_hedge = pd.DataFrame(event_data_hedge)

df_pl_fi, daily_pl_fi, df_pos_intraday_fi, daily_pos_fi = get_pl_pos(FS, gearing, cost_const, FPV, df_data, pos_curve, event_vol, data_date_time)
#daily_pl_fi.cumsum().plot()

df_pl_hedge, daily_pl_hedge, df_pos_intraday_hedge, daily_pos_hedge = get_pl_pos(FS, gearing, cost_const, FPV, df_data_hedge, -1*pos_curve, event_vol_hedge, data_date_time)
#daily_pl_hedge.cumsum().plot()

daily_pl_all = daily_pl_fi + daily_pl_hedge
#daily_pl_all.to_csv(inputcsvdir + '\\' + 'UK_RX_spread_pl.csv')

daily_pos_all = pd.concat([daily_pos_fi, daily_pos_hedge], axis=1)
#daily_pos_all.columns = ['G pos', 'RX pos']
#daily_pos_all.to_csv(inputcsvdir + '\\' + 'UK_RX_spread_pos.csv')

#PL analysis
btest = btestlib.Backtest(params)
IR = btest.calc_IR(daily_pl_all, annual_factor= 252)
```
