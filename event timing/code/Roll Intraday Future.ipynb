```python
#Load daily data and first notice day
contract = 'OAT'
inputcsvdir = 'E:\Intraday\FOAT'
daily_summary_df = pd.read_excel(inputcsvdir + '\\'+ contract +'_daily.xlsx', sheetname = 'daily data', header = 1, na_values = ['NA'])
trading_days = pd.DatetimeIndex(daily_summary_df['Date'])

first_notice_day_df = pd.read_excel(inputcsvdir + '\\'+ contract +'_daily.xlsx', sheetname = 'first notice day', na_values = ['NA'])
first_notice_day_df = first_notice_day_df.dropna(how = 'any')
first_notice_day_df['Contract'] = [x.split(' ')[0] for x in first_notice_day_df.ix[:,0]]
```

### Loop through all contracts to get Roll adjust data

```python
df = pd.DataFrame()
roll_day_shift = - 1
# get the first contract
contract_front = first_notice_day_df['Contract'][0]
df = pd.read_csv(inputcsvdir + '\\'+ contract_front +'.csv', na_values = ['NA'], parse_dates= False)
date_time = pd.to_datetime(df.Date + ' '+ df.Time, infer_datetime_format = True)
df.set_index(date_time, inplace=True)

for idx in range(1, len(first_notice_day_df)):
    i_contract = first_notice_day_df['Contract'].ix[idx]
    
    #check if file exist
    filename = inputcsvdir + '\\'+ i_contract +'.csv'
    if not(os.path.isfile(filename)):
        break
    
    #get the latest OHLC data    
    df_new = pd.read_csv(filename, na_values = ['NA'], parse_dates= False)
    date_time_new = pd.to_datetime(df_new.Date + ' '+ df_new.Time, infer_datetime_format = True)
    df_new.set_index(date_time_new, inplace=True)
    
    #get previous contract first notice day
    first_notice_i = first_notice_day_df['Date'].ix[idx - 1]
    roll_date = dttm.shift_by_weekdays(first_notice_i, roll_day_shift)
    before_roll_date = dttm.shift_by_weekdays(first_notice_i, roll_day_shift - 1)
    roll_date = dttm.get_closest_date(roll_date, trading_days, direction='backward')
    before_roll_date = dttm.get_closest_date(before_roll_date, trading_days, direction='backward')
    # convert to string
    roll_date_str = roll_date.strftime('%Y-%m-%d')
    before_roll_date_str = before_roll_date.strftime('%Y-%m-%d')
    
    #get back adjusted the difference
    close_st_time = '17:25'
    close_ed_time = '17:30'
    close_front =df.ix[before_roll_date_str].ix[:,'Close'].between_time(close_st_time, close_ed_time).median()
    close_back =df_new.ix[before_roll_date_str].ix[:,'Close'].between_time(close_st_time, close_ed_time).median()
    roll_adj = round(close_back -close_front, 2)
    #get the updated data
    df_update = et.update_roll_adjusted_intraday_series(df, df_new, roll_date_str, before_roll_date_str, roll_adj)
    df = df_update.ix[:, ['Symbol', 'Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume']] 
```

```python
df.to_csv(inputcsvdir + '\\'+ contract + '_2014.csv', index = False)
```
