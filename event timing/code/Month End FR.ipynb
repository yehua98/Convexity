```python
# plot mean function
def plot_data_mean(data_dict, contracts, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    ax1 = grid()
    hold(True)
    for i_contract in contracts:
        #convert to dataframe
        df_data = pd.DataFrame(data = data_dict[i_contract], index = index_dates, columns = column_times)
        df_data.ix[plot_date_st:plot_date_ed].mean(axis = 0).T.plot(ax = ax1, label = i_contract)
    for i in range(1,day_end - day_st):
        axvline(x = i* time_int_n-1, color = 'k')
    legend(loc = 'best') 
    return ax1
```

```python
def plot_pos_data(data_pos, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    #convert to dataframe
    df_data = pd.DataFrame(data = data_pos, index = index_dates, columns = column_times)
    ax1 = df_data.ix[plot_date_st:plot_date_ed].T.plot(grid = True)
    for i in range(1,day_end - day_st):
        axvline(x= i* time_int_n, color = 'k')
    legend(loc = 'best')
    return ax1
```

```python
def Event_pl_pos_output(Assets, data, col_name, filename):
    # save data to filename csv file
    df_data = pd.DataFrame()
    col_list= [x + y for x in Assets for y in col_name]
    for i in range(len(Assets)):
        df_data = pd.concat([df_data, data[i]], axis = 1)
    df_data.columns = col_list
    df_data.to_csv(filename)    
```

```python
#load main event data
contracts = ['RX', 'GX']
t_index =pd.date_range('2013-01-01 08:06', '2013-01-01 18:11', freq = '5min') 
time_zone = 'Europe/London'
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
day_st  = -2
day_end = 2
time_interval_str = [itime.strftime("%H:%M") for itime in time_interval*(day_end -day_st)]
event_vol_dict, df_variance_profile, event_data_dict, event_data_vol_adj_dict, event_data_date_time  = et.get_event_intraday_data_outright(inputcsvdir, contracts, time_interval, time_int_n, time_zone, day_st, day_end, Events_dates, len_events)
event_data_vol_adj_dict['RX + GX'] = 0.5*event_data_vol_adj_dict['RX'] + 0.5*event_data_vol_adj_dict['GX']
event_data_vol_adj_dict['RX - GX'] = 0.5*event_data_vol_adj_dict['RX'] - 0.5*event_data_vol_adj_dict['GX']
event_data_vol_adj_dict['RX biased'] = 0.7*event_data_vol_adj_dict['RX'] + 0.3*event_data_vol_adj_dict['GX']
event_data_vol_adj_dict['GX biased'] = 0.3*event_data_vol_adj_dict['RX'] + 0.7*event_data_vol_adj_dict['GX']
```

```python
# buy all to get the position
pos_curve_dict = {}
ewma_seedperiod = 12
window_size = 12
smooth_window = 24.0
mean = np.concatenate((np.diff(event_data_vol_adj_dict['RX + GX'],n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['RX'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['RX'] = 0.55 * pos_1 + 0.45 * pos_2   # pos_1 expanding window, pos_2 rolling window

mean = np.concatenate((np.diff(event_data_vol_adj_dict['RX biased'],n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['RX'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['RX biased'] = 0.55 * pos_1 + 0.45 * pos_2   # pos_1 expanding window, pos_2 rolling window

mean = np.concatenate((np.diff(event_data_vol_adj_dict['GX biased'],n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['GX'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['GX biased'] = 0.55 * pos_1 + 0.45 * pos_2   # pos_1 expanding window, pos_2 rolling window

#smooth_window = 24.0
#mean = np.concatenate((np.diff(event_data_vol_adj_dict['Z'],n=1,axis = 1), np.zeros((len_events, 1))), 1)
#variance = np.tile(df_variance_profile['Z'].values.T,(len_events,day_end - day_st))
#pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
#pos_curve_dict['Z'] = 1.0 * pos_1 + 0.0 * pos_2   # pos_1 expanding window, pos_2 rolling window
```
