```python
# plot mean function
def plot_data_mean(data_dict, contracts, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    ax1 = grid()
    hold(True)
    for i_contract in contracts:
        #convert to dataframe
        df_data = pd.DataFrame(data = data_dict[i_contract], index = index_dates, columns = column_times)
        df_data.ix[plot_date_st:plot_date_ed].mean(axis = 0).T.plot(ax = ax1, label = i_contract)
    for i in range(1,day_end - day_st):
        axvline(x = i* time_int_n-1, color = 'k')
    legend(loc = 'best') 
    return ax1
```

```python
def plot_pos_data(data_pos, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    #convert to dataframe
    df_data = pd.DataFrame(data = data_pos, index = index_dates, columns = column_times)
    ax1 = df_data.ix[plot_date_st:plot_date_ed].T.plot(grid = True)
    for i in range(1,day_end - day_st):
        axvline(x= i* time_int_n, color = 'k')
    legend(loc = 'best')
    return ax1
```

```python
def Event_pl_pos_output(Assets, data, col_name, filename):
    # save data to filename csv file
    df_data = pd.DataFrame()
    col_list= [x + y for x in Assets for y in col_name]
    for i in range(len(Assets)):
        df_data = pd.concat([df_data, data[i]], axis = 1)
    df_data.columns = col_list
    df_data.to_csv(filename)    
```

```python
UK = pd.read_csv(inputcsvdir +'//'+ 'UK_updated.csv', index_col= 0, parse_dates= True, infer_datetime_format= True)
UK = UK['2005-01-01':'2014-02-01']
UK['Date'] = [x.date() for x in UK.index]
UK['Week Day'] = [x.weekday() for x in UK.index]
UK['Day of Month'] = [x.day for x in UK.index]
Events_dates = set(UK['Date'].ix[(UK['Tenor'] >= 5).values])
Events_dates = list((Events_dates))
Events_dates = sort(Events_dates)
Week_days = [i_date.weekday() for i_date in Events_dates]
Event_date_index = pd.DatetimeIndex(Events_dates)
df_week_days = pd.DataFrame(Week_days, index = Event_date_index, columns= ['Week of the day'])
len_events = len(Events_dates)
len_events
```

```python
#load main event data
contracts = ['G']
t_index = pd.date_range('2013-01-01 08:06', '2013-01-01 17:11', freq = '5min')
time_zone = 'Europe/London'
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
day_st  = -2
day_end = 2
time_interval_str = [itime.strftime("%H:%M") for itime in time_interval*(day_end -day_st)]
event_vol_dict, df_variance_profile, event_data_dict, event_data_vol_adj_dict, event_data_date_time  = et.get_event_intraday_data_outright(inputcsvdir, contracts, time_interval, time_int_n, time_zone, day_st, day_end, Events_dates, len_events)
```
