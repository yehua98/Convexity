```python
# load all intraday csv file
contract = 'TP'

#df= pd.read_csv(inputcsvdir + '\\FV_Back_Adjust_Continuous.csv', na_values = ['NA'], parse_dates= False)
df= pd.read_csv(inputcsvdir + '\\'+ contract +'_2014.csv', na_values = ['NA'], parse_dates= False)
#df= pd.read_csv(inputcsvdir + '\\'+ contract+'_2014.csv', na_values = ['NA'], parse_dates= False)

#set the datatime index
date_time = pd.to_datetime(df.Date + ' '+ df.Time, infer_datetime_format = True)
df.set_index(date_time, inplace=True)

#set the starting dates for data 
df = df['2005-01-01':]
df_close = (df['High'] +df['Low'])/2
```

```python
#get JPY trading days
start_dt = datetime.datetime(2005, 1, 1)
#start_dt = datetime.datetime(2010, 2, 2)
end_dt = datetime.datetime(2014, 1, 31)
freq = 'B'   # frequency - business days
calendar = 'JPY' 
dttime = datetimelib.Datetime()
dt_range = dttime.get_business_days(start_dt = start_dt, end_dt = end_dt, freq = freq, calendar = calendar)
dates = np.unique(dt_range.date)
dates = sorted(dates)
#get datetime
t_index_1 =pd.date_range('2013-01-01 09:01', '2013-01-01 11:01', freq = '5min')
t_index_2 =pd.date_range('2013-01-01 12:31', '2013-01-01 15:01', freq = '5min') 
t_index = t_index_1 + t_index_2 
time_interval = [t.time() for t in t_index]
dt_index = pd.DatetimeIndex([datetime.datetime.combine(dates_i,time_j) for dates_i in dates for time_j in time_interval])
dt_index = sorted(dt_index)
# get 5 mins data
data = pd.DataFrame(df.Close, columns=['Close'])  # here only used close price
data = data.reindex(dt_index, method = 'ffill')
data['Rtn'] = data.Close.diff(-1)   # use -1 to get the predictive variance
#output to csv
data['Close'].to_csv( inputcsvdir + '\\'+ contract + '_5min.csv', header = ['Close'])
```

```python
#convert to multi index dataframe to represent intraday data
arrays = [[x.date() for x in data.index]
         ,[x.time() for x in data.index]]
tuples = zip(*arrays)
m_index = pd.MultiIndex.from_tuples(tuples, names=['date', 'time'])
data_mi = pd.DataFrame(data['Rtn'].values, index=m_index)

# get the intraday variance profile
data_1 = data_mi.unstack()
risk = risklib.Risk()
vol_halflife = 18
min_periods = 18
rm_risk = risk.generate_ewma_riskmodel(data_1, vol_halflife = vol_halflife, min_periods= min_periods)
#get intraday variance profile
Var=rm_risk.get('var')
Var = Var[date(2011,3, 1):]
var_ratio = np.sum(Var.values[:, 0:-1], axis = 1)/Var.values[: ,-1] 
variance_profile = median(Var.values, axis = 0)
variance_profile[-1] =np.sum(variance_profile[0:-1])/mean(var_ratio)
variance_profile = variance_profile/sum(variance_profile)
figure()
plot(time_interval, variance_profile, 'b.-')
grid()
S_variance_profile = pd.Series(data = variance_profile, index =time_interval)
S_variance_profile.to_csv(inputcsvdir + '\\'+ contract + '_intraday_variance_5min.csv', header = ['Var'])
```

```python
#get daily ATR by intraday data
def get_ATR(data_df, open_time, close_time):
    df = data_df[['Open', 'High', 'Low', 'Close']]
    df = df.between_time(open_time, close_time)
    trading_days = np.unique(df.index.date)
    trading_days = sorted(trading_days)
    #get daily OHLC
    df_days_OHLC = pd.DataFrame(np.nan, index = trading_days, columns=df.columns) 
    for i_day in trading_days:
        df_daily_data = df[i_day.strftime('%Y-%m-%d')]
        df_days_OHLC.ix[i_day, 'Open'] = df_daily_data.Open[0]
        df_days_OHLC.ix[i_day, 'High'] = np.max(df_daily_data.values)
        df_days_OHLC.ix[i_day, 'Low'] = np.min(df_daily_data.values)
        df_days_OHLC.ix[i_day, 'Close'] = df_daily_data.Close[-1]
    
    df_ATR = pd.DataFrame(np.nan, index = trading_days, columns=['ATR'])
    OHLC = df_days_OHLC.values
    
    #get daily ATR
    df_ATR.ATR[0] = OHLC[0, 1] - OHLC[0, 2]
    for i, i_day in enumerate(trading_days[1:], start = 1):
        df_ATR.ix[i_day, 'ATR'] = max([OHLC[i, 1]- OHLC[i, 2], abs(OHLC[i, 1] - OHLC[i-1, 3]),abs(OHLC[i, 2] - OHLC[i-1, 3])]) 
    
    #get smoothed ATR through EWMA
    df_ATR['ewma'] = pd.ewma(df_ATR.ATR, com = 2, min_periods=3)
    return df_ATR
```

```python
#get daily ATR
df_ATR  = get_ATR(df.reindex(dt_index, method = 'ffill'), open_time = '09:01',close_time = '15:01')
df_ATR.plot()
df_ATR.to_csv(inputcsvdir + '\\'+ contract + '_ATR.csv')
```
