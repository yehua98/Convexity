```python
def convert_list_2_series(bar_list):
    dt_index = []
    value = []
    for i in range(len(bar_list)):
        dt_index.append(bar_list[i]['column'])
        value.append(float(bar_list[i]['value']))
    dt_index = pd.DatetimeIndex(dt_index)
    s1 = pd.Series(data = value, index = dt_index)
    return s1
```

```python
def convert_to_datetime(df, col): 
    datetime_ix = [pd.to_datetime(str(int(x))) for x in df[0]]
    df_new = pd.DataFrame(data = df[1].values, index = datetime_ix)
    df_new.columns = [col]
    return df_new
```
