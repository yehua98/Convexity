```python
def aggregate_pl_pos(inputcsvdir, event_name, signal_list):
    df_pl = pd.DataFrame()
    df_pos = pd.DataFrame()
    for i in range(len(signal_list)):
        # pl
        pl_filename = inputcsvdir + '\\'+ event_name + signal_list[i] + '_intraday_pl.csv'
        df_pl_new = pd.read_csv(pl_filename, index_col=0, parse_dates=True, infer_datetime_format= True)
        df_pl = df_pl.add(df_pl_new, fill_value= 0.0)
        # pos
        pos_filename = inputcsvdir + '\\'+ event_name + signal_list[i] + '_intraday_pos.csv'
        df_pos_new = pd.read_csv(pos_filename, index_col=0, parse_dates=True, infer_datetime_format= True)
        df_pos = df_pos.add(df_pos_new, fill_value= 0.0)
        
    return df_pl, df_pos  
```

```python
# load 257 PL
event_name = 'USD257_'
signal_list = ['friday_consession', 'monday_short_squeeze', '2y_short_squeeze',  '5y_buyall', '7y_consession', '7y_buyall']
df_pl_257, df_pos_257 = aggregate_pl_pos(inputcsvdir, event_name, signal_list)
```

```python
# load 257 curve trade PL
event_name = 'USD257_curve_'
signal_list = ['previous_week_consession', 'after_5y_auction']
df_pl_257_curve, df_pos_257_curve = aggregate_pl_pos(inputcsvdir, event_name, signal_list)
```

```python
df_pl_257_all = df_pl_257.add(df_pl_257_curve, fill_value= 0.0) 
df_pos_257_all = df_pos_257.add(df_pos_257_curve, fill_value= 0.0) 
df_pl_257_all = df_pl_257_all/495019.0*150000
df_pos_257_all = df_pos_257_all/495019.0*150000
```

```python
#output PL for 257 event
df_pl_257_all.to_csv(inputcsvdir + '\\'+ 'USD257_all_intraday_pl.csv')
df_pos_257_all.to_csv(inputcsvdir + '\\'+ 'USD257_all_intraday_pos.csv')
```

```python
df_pl_257_daily = df_pl_257_all.ix[:, ['TY pl','FV pl','WN pl', 'ES pl']].sum(axis= 1).resample('B', how = 'sum')
df_pl_257_daily_net = df_pl_257_all.ix[:, ['TY pl_net','FV pl_net','WN pl_net','ES pl_net']].sum(axis= 1).resample('B', how = 'sum')
df_pl_257_daily = df_pl_257_daily.fillna(value = 0)
df_pl_257_daily_net = df_pl_257_daily_net.fillna(value = 0)
ax1 = df_pl_257_daily.cumsum().plot()
df_pl_257_daily_net.cumsum().plot(ax = ax1)
# monthly also per event PL
df_pl_257_monthly = df_pl_257_daily.resample('M', how = 'sum')
df_pl_257_monthly_net = df_pl_257_daily.resample('M', how = 'sum')
df_pl_257_monthly_net.mean()
```
