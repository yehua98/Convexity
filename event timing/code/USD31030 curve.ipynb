```python
# plot mean function
def plot_data_mean(data_dict, contracts, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    ax1 = grid()
    hold(True)
    for i_contract in contracts:
        #convert to dataframe
        df_data = pd.DataFrame(data = data_dict[i_contract], index = index_dates, columns = column_times)
        df_data.ix[plot_date_st:plot_date_ed].mean(axis = 0).T.plot(ax = ax1, label = i_contract)
    for i in range(1,day_end - day_st):
        axvline(x = i* time_int_n-1, color = 'k')
    legend(loc = 'best') 
    return ax1
```

```python
def plot_pos_data(data_pos, index_dates, column_times, plot_date_st, plot_date_ed, day_st, day_end, time_int_n):
    rcParams['figure.figsize'] = 18, 3
    #convert to dataframe
    df_data = pd.DataFrame(data = data_pos, index = index_dates, columns = column_times)
    ax1 = df_data.ix[plot_date_st:plot_date_ed].T.plot(grid = True)
    for i in range(1,day_end - day_st):
        axvline(x= i* time_int_n, color = 'k')
    legend(loc = 'best')
    return ax1
```

```python
def Event_pl_pos_output(Assets, data, col_name, filename):
    # save data to filename csv file
    df_data = pd.DataFrame()
    col_list= [x + y for x in Assets for y in col_name]
    for i in range(len(Assets)):
        df_data = pd.concat([df_data, data[i]], axis = 1)
    df_data.columns = col_list
    df_data.to_csv(filename)    
```

## 30 year, WN_FV spread rally into Auction. good performance before 2013 but flat since then. Cost is also an issue

```python
#load main event data
contracts = ['FV', 'TY', 'US', 'WN', 'ES']
t_index = pd.date_range('2013-01-01 00:01', '2013-01-01 14:56', freq = '5min')
time_zone = 'US/Central'
time_interval = [t.time() for t in t_index]
time_int_n = len(time_interval)
day_st  = 0
day_end = 2
time_interval_str = [itime.strftime("%H:%M") for itime in time_interval*(day_end -day_st)]

event_vol_dict, df_variance_profile, event_data_dict, event_data_vol_adj_dict, event_data_date_time  = et.get_event_intraday_data_outright(inputcsvdir, contracts, time_interval, time_int_n, time_zone, day_st, day_end, Events_dates, len_events)
event_data_vol_adj_dict['WN_FV'] = event_data_vol_adj_dict['WN'] - event_data_vol_adj_dict['FV']
event_data_vol_adj_dict['WN_TY'] = event_data_vol_adj_dict['WN'] - event_data_vol_adj_dict['TY'] 
```

```python
ax1 = plot_data_mean(event_data_vol_adj_dict, ['WN_FV', 'WN_TY'], Event_date_index, time_interval_str, '2010', '2014', day_st, day_end, time_int_n)
```

```python
pos_curve_dict = {}
ewma_seedperiod = 10
window_size = 10
smooth_window = 24.0
mean = np.concatenate((np.diff(event_data_vol_adj_dict['WN_FV'],n=1,axis = 1), np.zeros((len_events, 1))), 1)
variance = np.tile(df_variance_profile['WN'].values.T,(len_events,day_end - day_st))
pos_1, pos_2 = et.generate_pos_curve(mean, variance, ewma_seedperiod, window_size, smooth_window, 0*time_int_n, (day_end - day_st)*time_int_n)
pos_curve_dict['30y rally'] = 1.0 * pos_1 + 0.0 * pos_2   # pos_1 expanding window, pos_2 rolling window
```

```python
#ax1 = plot_pos_data(pos_1, Event_date_index, time_interval_str, '2013', '2014', day_st, day_end, time_int_n)  
ax2 = plot_pos_data(pos_1, Event_date_index, range(len(time_interval_str)), '2013', '2013', day_st, day_end, time_int_n)
legend(loc= 'center left')
ax2 = plot_pos_data(pos_2, Event_date_index, time_interval_str, '2013', '2013', day_st, day_end, time_int_n) 
legend(loc = 'center left')
```

```python
FS = 1000000
gearing = 3.64   # mean(PL) = 60K
FPV_dict = {'FV': 1000, 'TY': 1000, 'US': 1000, 'WN': 1000, 'ES': 50}
cost_const_dict = {'FV': 0.5*1.0/64, 'TY': 0.5*1.0/64, 'US': 0.5*1.0/32, 'WN': 0.5*1.0/32, 'ES': 0.5*0.25}
Asset = 'WN'
FPV = FPV_dict['WN']
cost_const = 0.5 *cost_const_dict['WN']
df_data = pd.DataFrame(event_data_dict['WN'])
#event_vol = spread_vol
event_vol = event_vol_dict['WN']
pos_curve = pos_curve_dict['30y rally'].copy()

pos_curve[:, :80] = 0.0
pos_curve[:, 80:140] = 1.0
pos_curve[:, 141:] = 0.0

df_pl_fi_intraday, daily_pl_fi, df_pos_intraday_fi, daily_pos_fi, df_pl_fi_events, df_pl_fi_events_net = et.get_pl_pos(FS, gearing, cost_const, FPV, df_data, pos_curve, event_vol, Events_dates, event_data_date_time)
daily_pl_fi.cumsum().plot()

Hedge_Asset = 'TY' 
FPV_hedge = FPV_dict['TY']
cost_const_hedge = 0.5*cost_const_dict['TY']
df_data_hedge = pd.DataFrame(event_data_dict['TY'])
#event_vol_hedge = spread_vol
event_vol_hedge = event_vol_dict['TY']
pos_curve_hedge = - pos_curve.copy()
#pos_curve_hedge = - pos_curve.copy() * spread_ratio[:, np.newaxis]

df_pl_hedge_intraday, daily_pl_hedge, df_pos_intraday_hedge, daily_pos_hedge, df_pl_hedge_events, df_pl_hedge_events_net  = et.get_pl_pos(FS, gearing, cost_const_hedge, FPV_hedge, df_data_hedge, pos_curve_hedge, event_vol_hedge, Events_dates, event_data_date_time)
daily_pl_hedge.cumsum().plot()

daily_pl_all = daily_pl_fi + daily_pl_hedge
daily_pl_all.cumsum().plot()

#PL analysis
btest = btestlib.Backtest(params)
IR = btest.calc_IR(daily_pl_all.ix['2010':], annual_factor= 252)
hit_ratio = et.get_hit_ratio(daily_pl_all.ix['2010':]['pl'].values)
hit_ratio_net = et.get_hit_ratio(daily_pl_all.ix['2010':]['pl_net'].values)
event_hit_ratio = et.get_hit_ratio((df_pl_fi_events_net.sum(axis = 1) + df_pl_hedge_events_net.sum(axis = 1)).ix['2010':].values)
IR, hit_ratio, hit_ratio_net, event_hit_ratio
#gearing/50000.0*(df_pl_fi_events_net.sum(axis = 1) + df_pl_hedge_events_net.sum(axis = 1)).mean()
```
